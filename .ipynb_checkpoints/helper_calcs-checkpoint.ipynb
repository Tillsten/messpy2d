```python
def phase_ifg(mr, step, j=2):
    #mr = np.hstack((mr, np.zeros_like(mr)))
    mr_spec = np.fft.fft(mr, mr.size)[:mr.size//2]
    fr = np.fft.fftfreq(mr_spec.size*2, step)*1e15
    wn = (fr/speed_of_light)/1e2
    wk  = wn[:mr.size//2]
    mr_absspec = np.abs(mr_spec)
    pp = np.argmax(mr_absspec[50:])+50
    sl = slice(pp-10, pp+10)
    idx = mr_absspec[sl] > 35
    
    mr_phase = np.unwrap(np.angle(mr_spec[sl][idx]))
    #mr_phase = np.angle(mr_spec)

    #A = np.vstack([wk[sl][idx]-wk[pp], (wk[sl][idx]-wk[pp])**3, np.ones(wk[sl][idx].size)]).T
    #deriv, _, const = np.linalg.lstsq(A, mr_phase)[0]
    A = np.vstack([wk[sl][idx]-wk[pp], np.ones(wk[sl][idx].size)]).T
    deriv, const = np.linalg.lstsq(A, mr_phase)[0]
    const =  (( -const + np.pi) % (2.0 * np.pi ) - np.pi)*-1 
    #plt.plot(wk, mr_absspec)
    #plt.twinx()
    #plt.plot(wk, np.angle(mr_spec))
    #plt.plot(wk[sl][idx], np.angle(mr_spec)[sl][idx])
    #plt.xlim(2200, 2400)
    return const, deriv
binned_pyro = m.copy()/m.max()
m_argmax = np.argmax(abs(binned_pyro)[10:])
mr = np.roll(binned_pyro, -m_argmax)
phase_ifg(mr, step=step)
```
